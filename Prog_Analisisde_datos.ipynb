```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

base_url = "https://www.football-espana.net"
url = base_url + "/category/la-liga/real-madrid"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# 🔹 Función para obtener contenido del artículo
def get_article_text(url):
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")

        paragraphs = soup.find_all("p")
        text = " ".join([p.get_text() for p in paragraphs])

        return text
    except:
        return None

# 🔹 Obtener página principal
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

posts = soup.find_all("div", class_="post-container")

data = []

for post in posts[:10]:  # usa 10 para evitar bloqueos

    # título
    title_tag = post.find("a", title=True)
    title = title_tag.get("title") if title_tag else None

    # link
    link_tag = post.find("a", href=True)
    link = link_tag.get("href") if link_tag else None

    # convertir a URL completa
    if link and link.startswith("/"):
        link = base_url + link

    # 🔥 contenido del artículo
    content = get_article_text(link) if link else None

    data.append({
        "title": title,
        "content": content,
        "url": link
    })

    # ⚠️ pausa para no saturar el servidor
    time.sleep(1)

# Mostrar resultados
for d in data:
    print(d)

# Guardar en CSV
df = pd.DataFrame(data)
df.to_csv("real_madrid_articles.csv", index=False)

print("\n✅ Datos guardados en real_madrid_articles.csv")
import pandas as pd

df = pd.DataFrame(data)

# Mostrar las primeras filas
df.head()
```
